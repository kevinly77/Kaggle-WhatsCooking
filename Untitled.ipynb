```python
from collections import Counter
counters = {}
for cuisine in train_data['cuisine'].unique():
    counters[cuisine] = Counter()
    indices = (train_data['cuisine'] == cuisine)
    for ingredients in train_data[indices]['ingredients']:
        counters[cuisine].update(ingredients)


```

```python
top10 = pd.DataFrame([[items[0] for items in counters[cuisine].most_common(10)] for cuisine in counters],
            index=[cuisine for cuisine in counters],
            columns=['top{}'.format(i) for i in range(1, 11)])
top10

```

```python
#train_data['all_ingredients'] = train_data['ingredients']
#train_data['all_ingredients'] = train_data['all_ingredients'].str.split(",")




train_data['all_ingredients'] = train_data['ingredients'].map(','.join)

#rain_data['all_ingredients'] = train_data['all_ingredients'].apply(remove_sw)


                                                        

#train_data['all_ingredients'].values
train_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words=['fresh','cracked'])
cv.fit(train_data['all_ingredients'].values)

X = cv.transform(train_data['all_ingredients'].values)

X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
from keras.utils.np_utils import to_categorical




y = enc.fit_transform(train_data['cuisine'])
y2 = to_categorical(y)
#y[:100]
y2
```

```python
from keras import models
from keras import layers
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y2, test_size=0.2)


model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape = (2866,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(20, activation='softmax'))

x_val = X_train[:1000]
partial_x_train = X_train[1000:]

y_val = y_train[:1000]
partial_y_train = y_train[1000:]
print(partial_y_train.shape)
print(partial_y_train)

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(X_train,
            y_train,
            epochs=20,
            batch_size=512)




```

```python
test
pred_test = test
pred_test['all_ingredients'] = test['ingredients'].map(','.join)
#pred_test['ingredients'] = cv.transform(pred_test['all_ingredients'].values)
z_test = cv.transform(pred_test['all_ingredients'].values)
z_test.shape




#train_data['all_ingredients'] = train_data['ingredients'].map(','.join)
#from sklearn.feature_extraction.text import CountVectorizer
#cv = CountVectorizer(stop_words=['fresh','cracked'])
#cv.fit(train_data['all_ingredients'].values)

#X = cv.transform(train_data['all_ingredients'].values)

#X.shape
```

```python
prediction_test = model.predict(z_test).argmax(axis=1)
result = enc.inverse_transform(prediction_test)
```
