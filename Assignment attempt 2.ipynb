```python
import keras
keras.__version__
from keras import models
from keras import layers


import numpy as np
import pandas as pd
```

```python
#train = pd.read_csv('C:/Users/Kevin/Desktop/Machine Learning/Assignment_2/train.json')
train = pd.read_json('train.json')
test = pd.read_json('test.json')

#20 unique classes
train['cuisine'].nunique()
```

```python
train['ingredients']
```

```python
from sklearn.feature_extraction import text
my_additional_stop_word_list = ['ground','fresh','free','kosher','powder','finely','fat-free','reduced','all-purpose','light','crumbs','freshly','skimmed','crusty','boiling','chopped','unsalted','whole','extra-virgin','cracked','flat','leaf','smoked','shredded','frozen','in','uncooked','crushed','sliced','fat','skimmed','dried','granulated','dark','condensed','of','skinless','large','diced','spicy','lean','grated','caps','dried','cooked','shanghai','powdered','boneless','purple','yellow','sodium','red','leav','clarified','unflavored','with','english','part-skin','low','cut','into','loaf','active','dry','paced','loosely','packed','1%','2%','3%','sweetened','coarse','drained','store','bought','KRAFT','Kraft','Zesty','Grated','crumbles','brine','cured','self','rising','blanched','Hidden','Valley®','Original','Ranch®','Dip','Mix','Lipton®','Brew','Family','Size','Tea','Bags','extra','wide','firm']
stop_wordss = text.ENGLISH_STOP_WORDS.union(my_additional_stop_word_list)
stop_wordss
```

```python
train['vectorized_ingredients'] = train['ingredients'].map(','.join)
train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words =stop_wordss, lowercase=True, strip_accents='unicode')
```

```python
cv.fit(train['vectorized_ingredients'].values)
```

```python
X = cv.transform(train['vectorized_ingredients'].values)
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils.np_utils import to_categorical
enc = LabelEncoder()
y = enc.fit_transform(train['cuisine'])
y2 = to_categorical(y)
y2
```

```python
from keras import models
from keras import layers
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y2, test_size=0.3)
```

```python
model = models.Sequential()
model.add(layers.Dense(800, activation='relu', input_shape = (2757,))),
model.add(layers.Dense(1000, activation='relu')),
model.add(layers.Dropout(0.50)),
model.add(layers.Dense(20, activation='softmax'))
```

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=40, batch_size=512)

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=20, batch_size=1024)
```

```python
pred_test = test
pred_test['ingredients'] = test['ingredients'].map(','.join)
```

```python
z_test = cv.transform(pred_test['ingredients'].values)
```

```python
z_test.shape
```

```python
prediction_test = model.predict(z_test).argmax(axis=1)
result = enc.inverse_transform(prediction_test)
```

```python
result
```

```python
submission = pd.DataFrame()
submission['id'] = test['id']
submission['cuisine'] = result
submission.to_csv('submission_32.csv', index=False)
```
